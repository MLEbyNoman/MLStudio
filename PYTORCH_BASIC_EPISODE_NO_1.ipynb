# **Tensor DataTypes** (Think of bits like digits in a number:

16-bit → short number (less detail)

32-bit → normal number (good detail)

64-bit → very long number (very high detail)

More bits = more accuracy and bigger rang)

```python
sample = torch.rand(3,4)
sample
```

```python
sample.dtype
```

```python
sample=torch.arange(3.,4.,dtype=int)
sample,sample.dtype
```

# ***Change Tensor DataType***

```python
x = torch.arange(1.,10.)
x,x.dtype
```

```python
tensor_float16=x.type(torch.float16)
tensor_float16
```

```python
tensor_int8=x.type(torch.int8)
tensor_int8
```
